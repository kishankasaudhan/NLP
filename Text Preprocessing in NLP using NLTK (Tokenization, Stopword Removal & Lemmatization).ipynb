```python
sentences = [
    "i am learning python and python",
    "we are happy today",
    "i am sad today"
]
```

```python
#Tokenization
import nltk
import string
from nltk.tokenize import word_tokenize
#stopwords list
from nltk.corpus import stopwords
#Stemming
from nltk.stem import PorterStemmer
#Lemmatization
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
def preprocess_text(documents):
    english_stopwords = stopwords.words("english")
    punctuations = string.punctuation
    cleaned_documents = []

    for doc in documents:
        # Step 1: Lowercase
        raw_text = doc.lower()

        # Step 2: Tokenization
        tokens = word_tokenize(raw_text)

        # Step 3: Remove stopwords
        filtered_tokens = [word for word in tokens if word not in english_stopwords]

        # Step 4: Remove punctuation
        clean_tokens = [word for word in filtered_tokens if word not in punctuations]

        # Step 5: Lemmatization
        wnet = WordNetLemmatizer()
        lemmatized_words = [wnet.lemmatize(word, "v") for word in clean_tokens]

        # Step 6: Keep only alphabetic words
        final_tokens = [word for word in lemmatized_words if word.isalpha()]

        cleaned_text = " ".join(final_tokens)
        cleaned_documents.append(cleaned_text)

    return cleaned_documents

output = preprocess_text(sentences)
print(output)
```
